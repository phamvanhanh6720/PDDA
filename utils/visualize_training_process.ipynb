```python
def get_log(file_log):

    with open(file_log, 'r') as file:
        text = file.read()
        lines = text.split('\n')
        lines = lines[:-1]

    train_auc = list()
    train_ap = list()

    test_auc = list()
    test_ap = list()

    for i in range(1, len(lines), 2):
        line = lines[i]
        info = [float(_) for _ in line.split(' ') if _ != '']
        train_auc.append(info[1])
        train_ap.append(info[2])

        test_auc.append(info[4])
        test_ap.append(info[5])

    return train_auc, train_ap, test_auc, test_ap
```

```python
def visualize(train_res, test_res, title):
    import matplotlib.pyplot as plt

    plt.figure(figsize=(10, 6))
    x = [i+1 for i in range(len(train_res))]
    plt.plot(x, train_res, label='train')
    plt.plot(x, test_res, label='test')
    plt.title(title)
    plt.grid(True)
    plt.legend()
```

```python
file_log = '../weight/5_fold/1th.txt'
```

```python
train_auc, train_ap, test_auc, test_ap = get_log(file_log=file_log)
```

```python
visualize(train_auc, test_auc, 'AUC')
```

```python
visualize(train_ap, test_ap, 'AP')
```
